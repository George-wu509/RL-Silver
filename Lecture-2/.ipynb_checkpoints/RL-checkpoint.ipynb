We can now sample what we'll call "episodes" further on: sequences of states we visit before getting to sleep.



````python
world = {
    0: 'Class 1',
    1: 'Class 2',
    2: 'Class 3',
    3: 'Pass',
    4: 'Pub',
    5: 'Facebook',
    6: 'Sleep'
}
````

````python
# Transition matrix for our Markov chain

P = np.array([
        [0,.5,0,0,0,.5,0], # C1
        [0,0,.8,0,0,0,.2], # C2
        [0,0,0,.6,.4,0,0], # C3
        [0,0,0,0,0,0,1.], # Pass
        [.2,.4,.4,0,0,0,0], # Pub
        [.1,0,0,0,0,.9,0], # Facebook
        [0,0,0,0,0,0,1] # Sleep
    ])
````

````python
episodes = []
while len(episodes) < 3:
    state = 0
    experience = []
    while state != 6: # while not asleep
        experience.append(state)
        state = np.random.choice(np.arange(7), size=1, p=P[state,:].ravel())[0]
    episodes.append(experience)
````

````python
for ep in episodes: 
    print [world[i] for i in ep] 
    print 
````

Recall that **Markov Reward Process** is defined as tuple: $<\mathcal{S}, \mathcal{P}, \mathcal{R}, \gamma>$

$\mathcal{R}_s = \mathbb{E}[R_{t+1}\ |\ S_t = s]$

$\gamma \in [0,1]$

````python
R = np.array([-2, -2, -2, 10, 1, -1, 0])
````

Recall that value function for a given state is defined as:
$v(s) = \mathbb{E}[G_t\ |\ S_t = s]$

where

$G_t = R_{t+1} + \gamma R_{t+2} + ... = \sum_{k=0}^{\infty}\gamma^k R_{t+k+1}$

so we can easily compute that by naively running many episodes and averaging the returns $G_t$

**Note**: the slides use incorrect implementation to compute values printed in the graph. Or if not incorrect, then at least incosistent with equaions at slide 15

Slides use:

Since the slide **15** says $G_1 = R_2 + \gamma R_3 + ...$, not $G_1 = R_1 + \gamma R_2 + ... $

````python
from numba import jit, int64, float64
gamma = 0.

@jit(float64[:](float64[:,:], float64[:,:], int64, int64, float64))
def get_value_func(P, R, num_episodes=10000, episode_len=100, gamma=0.):
    num_states = P.shape[0]
    returns = np.zeros(R.shape)
    
    for start_state in range(num_states): # for each state
        EG = 0
        
        for i in range(num_episodes): # run 10000 simulations
            state = start_state
            G = 0
            
            for j in range(episode_len): # run a 100 steps 
                G += (gamma**j) * R[state]
                state = np.random.choice(np.arange(7), size=1, p=P[state,:].ravel())[0]
                
            EG += G
            
        EG /= num_episodes
        returns[start_state] = EG
        
    return returns
````

````python
v = get_value_func(P, R, 10000, 3, 0.)
````

````python
for i in range(7): print "V[ {} ]\t=\t{}".format(world[i], v[i])
````

````python
v = get_value_func(P, R, 5000, 200, 0.9) # 0.9 ** 200 = 7e-10
````

````python
for i in range(7): print "V[ {} ]\t=\t{}".format(world[i], v[i])
````
